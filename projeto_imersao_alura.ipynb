```python
# !pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata

# Acesse a API Key armazenada no segredo "gemini"
api_key = userdata.get('gemini')

# Configure a API do Google Gemini apenas uma vez
genai.configure(api_key=api_key)

# Função para realizar o teste e retornar as áreas de maior aptidão
def realizar_teste():
  areas = {'a': 0, 'b': 0, 'c': 0, 'd': 0}

  print("**Seção 1: Lógica e Resolução de Problemas**")
  respostas = input("""
  1.  Ao se deparar com um problema complexo, você:
      a) Divide o problema em partes menores e busca soluções individuais.
      b)  Testa diferentes abordagens até encontrar uma que funcione.
      c) Analisa o problema a fundo para entender sua causa raiz antes de buscar soluções.
      d) Prefere pedir ajuda a alguém com mais experiência na área.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  2.  Qual dos seguintes jogos te atrai mais?
      a) Sudoku
      b) Xadrez
      c) Quebra-cabeças
      d) Jogos de construção (Minecraft, Lego, etc.)
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  3.  Você se considera uma pessoa:
      a) Criativa e com facilidade para encontrar soluções inovadoras.
      b)  Metódica e organizada, preferindo seguir processos já estabelecidos.
      c) Curiosa e investigativa, sempre em busca de novos conhecimentos.
      d) Colaborativa e com facilidade para trabalhar em equipe.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  print("\n**Seção 2:  Tecnologia e Ferramentas**")
  respostas = input("""
  1.  Qual seu nível de familiaridade com linguagens de programação?
      a) Conheço os conceitos básicos e consigo escrever pequenos programas.
      b) Domino pelo menos uma linguagem e consigo desenvolver projetos de média complexidade.
      c) Sou especialista em programação e consigo desenvolver projetos complexos em diferentes linguagens.
      d) Não possuo conhecimento em programação.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  2.  Você se interessa por:
      a)  Desenvolvimento de aplicativos e softwares.
      b)  Gerenciamento e configuração de redes de computadores.
      c)  Análise de dados e construção de modelos preditivos.
      d)  Configuração e manutenção de servidores e infraestrutura de TI.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  3.  Qual a sua ferramenta de busca preferida?
      a)  Google
      b)  DuckDuckGo
      c)  Bing
      d)  Ecosia
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  4.  Você utiliza algum sistema de controle de versão (Git, SVN)?
      a)  Sim, utilizo frequentemente.
      b)  Tenho conhecimento básico, mas não utilizo com frequência.
      c)  Já ouvi falar, mas nunca utilizei.
      d)  Não sei o que é um sistema de controle de versão.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  print("\n**Seção 3: Áreas de Interesse**")
  respostas = input("""
  1.  O que mais te fascina no mundo da tecnologia?
      a) A capacidade de criar soluções inovadoras para problemas reais.
      b) A possibilidade de conectar pessoas e informações através de redes complexas.
      c) A habilidade de extrair conhecimento a partir de grandes volumes de dados.
      d) A garantia de funcionamento e segurança da infraestrutura tecnológica.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  2.  Qual das áreas abaixo te desperta maior interesse?
      a) Desenvolvimento Web
      b) Cibersegurança
      c) Inteligência Artificial
      d) Cloud Computing
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  respostas = input("""
  3.  Quais perfis você costuma seguir nas redes sociais?
      a) Programadores, desenvolvedores e entusiastas de tecnologia.
      b)  Especialistas em segurança da informação e hackers éticos.
      c)  Cientistas de dados, estatísticos e matemáticos.
      d)  Administradores de sistemas e especialistas em infraestrutura.
  Resposta (a/b/c/d): """)
  areas[respostas.lower()] += 1

  maior_pontuacao = max(areas, key=areas.get)
  areas_de_aptidao = []

  for area, pontuacao in areas.items():
    if pontuacao == areas[maior_pontuacao]:
      if area == 'a':
        areas_de_aptidao.append("Programação")
      elif area == 'b':
        areas_de_aptidao.append("Redes")
      elif area == 'c':
        areas_de_aptidao.append("Ciência de Dados")
      elif area == 'd':
        areas_de_aptidao.append("Infraestrutura")

  return areas_de_aptidao

# Função para gerar sugestões de cursos com base nas áreas de aptidão
def gerar_sugestoes_cursos(areas_de_aptidao):
  prompt = f"Com base nas áreas de aptidão: {', '.join(areas_de_aptidao)}, sugira cursos da Alura, fornecendo o nome completo do curso e o link para ele."
  model = genai.GenerativeModel('gemini-1.5-pro-latest')
  response = model.generate_content([{'text': prompt}])
  return response.text

# --- Fluxo principal ---
areas = realizar_teste()

# Imprime a mensagem com a área de aptidão
for area in areas:
  print(f"\nParabéns, você tem uma aptidão para {area}!")

sugestoes = gerar_sugestoes_cursos(areas)
print(f"\nSugestões de cursos da Alura:\n{sugestoes}")
```
